```python
import json
#function to load data
def load_data(filename):
    with open(filename,"r") as f:
        return json.load(f)
#fuction to pages suggestion
def pages_you_make_like(user_id,data):
    liked_pages = {}
    for user in data['users']:
        liked_pages[user['id']] = set(user['liked_pages'])
    if user_id not in liked_pages:
        return []
    user_liked = liked_pages[user_id]
    page_suggestion = {}
    for other_user, pages in liked_pages.items():
        if other_user != user_id:
            shared_pages = user_liked.intersection(pages)
        for page in pages:
            if page not in user_liked:
                page_suggestion[page] = page_suggestion.get(page,0)+len(shared_pages)
    sorted_pages = sorted(page_suggestion.items(), key=lambda x:x[1],reverse=True)
    return [page_id for page_id,_ in sorted_pages]
data = load_data("massive_data.json")
user_id = 1
result = pages_you_make_like(user_id,data)
print(result)
```
